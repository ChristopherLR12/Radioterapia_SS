```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl

def truncate(number: float, max_decimals: int) -> float:
    int_part, dec_part = str(number).split(".")
    return float(".".join((int_part, dec_part[:max_decimals])))
```

## 15X

```python
TB_15X
```

```python
VB_15X
```

```python
er_15X = np.zeros((len(A_15X),len(A_15X)))

for i in np.linspace(0,len(A_15X)-1,len(A_15X),dtype=int):
    for j in np.linspace(0,len(A_15X)-1,len(A_15X),dtype=int):
        er_15X[i][j] = truncate(((np.abs(TB_15X[i][j]-VB_15X[i][j]))/(TB_15X[i][j]))*100,3)
        
np.mean(er_15X)
```

```python
tc = ["3","5","7","10","15","20","30","40"]

fig, ax = plt.subplots(figsize=(8,6))
im = ax.imshow(er_15X,cmap="Blues")


ax.set_xticks(np.arange(len(tc)), labels=tc, fontsize=13)
ax.set_yticks(np.arange(len(tc)), labels=tc, fontsize=13)

ax.spines[:].set_visible(False)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

ax.set_title("Comparación 15X",size=16)
cb = fig.colorbar(im, ax=ax)
cb.set_label("Error relativo [%]", rotation=90,size=14)
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
fig.tight_layout()
plt.savefig("15X_hm1.pdf",bbox_inches="tight")
```
