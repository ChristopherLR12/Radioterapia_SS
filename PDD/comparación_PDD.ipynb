```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
tc = ["040","060","080","100","120","150","200","250","300","350","400"]
```

```python
os.chdir(r"C:\Users\c-lop\Documents\GitHub\Radioterapia_SS\PDD\PDD_GD")

GD = {}

for i in range(len(tc)):
    nom = f"tamaño de campo_{tc[i]}x{tc[i]}.csv"
    GD[f"GD_{tc[i]}"] = pd.read_csv(nom, index_col=0)

print(GD.keys())
```

```python
os.chdir(r"C:\Users\c-lop\Documents\GitHub\Radioterapia_SS\PDD\PDD_AAA13")

AAA_13 = {}

for i in range(len(tc)):
    nom = f"tamaño de campo_{tc[i]}x{tc[i]}.csv"
    AAA_13[f"AAA_13_{tc[i]}"] = pd.read_csv(nom, index_col=0)

print(AAA_13.keys())
```

```python
os.chdir(r"C:\Users\c-lop\Documents\GitHub\Radioterapia_SS\PDD\PDD_AAA16")

AAA_16 = {}

for i in range(len(tc)):
    nom = f"tamaño de campo_{tc[i]}x{tc[i]}.csv"
    AAA_16[f"AAA_16_{tc[i]}"] = pd.read_csv(nom, index_col=0)

print(AAA_16.keys())
```

```python
os.chdir(r"C:\Users\c-lop\Documents\GitHub\Radioterapia_SS\PDD")
```

```python
def PDD_comp_GD(B, M, TC):
    error_por = (abs(B - M) / B) * 100

    prom_error = np.mean(error_por)
    std_error = float(error_por.std())

    text1 = f"{round(prom_error, 2)} ± {round(std_error, 2)}"

    plt.figure(figsize=(10, 8))
    plt.plot(B, lw=2.5, color="black", label="GD", ls=':', alpha=0.7)
    plt.plot(M, lw=2.5, color="navy", label="AAA_16", alpha=0.7)
    plt.plot(error_por, lw=2, color="red", label="Error porcentual relativo")
    plt.title(f"PDD AAA_16 vs GD / TC = {TC} x {TC} mm", size=20)
    plt.legend(fontsize=15)
    plt.xlabel("Profundidad [mm]", size=15)
    plt.ylabel("%", size=15)
    plt.grid()
    plt.text(50, 20, text1, c="red", size=18)
    plt.savefig(f"AAA_16_GD_{TC}.jpg", bbox_inches="tight", dpi=500)
```

```python
for i in range(len(tc)):
    PDD_comp_GD(GD[f"GD_{tc[i]}"], AAA_16[f"AAA_16_{tc[i]}"], tc[i])
```
