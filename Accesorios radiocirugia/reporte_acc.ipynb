```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd  
```

```python
data = pd.read_csv(r"C:\Users\c-lop\Documents\GitHub\Radioterapia_SS\Accesorios\colocacion_pre_radiocirugia.csv")
```

```python
data.describe()
```

```python
#Tipo paciente
sns.set(rc={'figure.figsize':(8,6)})
ax = sns.histplot(data["tipo_paciente"], color = "darkblue", edgecolor = "black", alpha = 0.75)

for i in ax.containers:
    ax.bar_label(i,)
    
plt.title("Tipo de paciente",size=18,fontname="DejaVu Sans",fontweight="bold")
plt.xlabel("Tipo",size=14,fontname="DejaVu Sans")
plt.ylabel("#",size=14,fontname="DejaVu Sans")
plt.xticks(size=20)
#plt.show()
plt.savefig("tipo_paciente.svg",bbox_inches="tight")
```

```python
#Caracteristicas paciente
sns.set(rc={'figure.figsize':(8,6)})
ax = sns.histplot(data["caracteristicas_paciente"], color = "darkblue", edgecolor = "black", alpha = 0.75)

for i in ax.containers:
    ax.bar_label(i,)
    
plt.title("Caractersticas del paciente",size=18,fontname="DejaVu Sans",fontweight="bold")
plt.xlabel("Caracterstica",size=14,fontname="DejaVu Sans")
plt.ylabel("#",size=14,fontname="DejaVu Sans")
plt.xticks(size=20,rotation='vertical')
#plt.show()
plt.savefig("carac_paciente.svg",bbox_inches="tight")
```

```python
#Tiempo de realizacin
sns.set(rc={'figure.figsize':(8,6)})
mean_norm = np.mean(data["tiempo_realizacion"])

#plt.axvline(data["tiempo_realizacion"], color="black", linestyle="--", label=f"Promedio: {mean_norm:.2f} %")
plt.plot(data["numero_de_realizacion"],data["tiempo_realizacion"],color = "navy", linewidth= 1.5, alpha = 1, marker = "o", mfc = "k", ms = 7)
plt.title("Tiempo de realizacin",size=18,fontname="DejaVu Sans",fontweight="bold")
plt.xlabel("Nmero de prctica",size=14,fontname="DejaVu Sans")
plt.ylabel("Tiempo [min]",size=14,fontname="DejaVu Sans")
#plt.show()
plt.savefig("tiempo_real.svg",bbox_inches="tight")
```
